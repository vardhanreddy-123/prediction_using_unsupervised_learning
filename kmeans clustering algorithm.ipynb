# KMeans clustering algorithm

we can use the KMeans clustering algorithm to predict the optimum number of clusters for the iris dataset and visualize it using Python.

Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

Loading the iris dataset

```python
iris = load_iris()
X = iris.data
```

Finding the optimum number of clusters for k-means classification

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

Plotting the results onto a line graph, allowing us to observe 'The elbow'

```python
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

From the above graph, we can see that the optimum number of clusters is 3, which is the point of the elbow in the graph.

Now, we can use the KMeans algorithm with n_clusters=3 to cluster the iris dataset.

Applying kmeans to the dataset / Creating the kmeans classifier

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)
```

Visualising the clusters

```python
plt.scatter(X[pred_y == 0, 0], X[pred_y == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[pred_y == 1, 0], X[pred_y == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[pred_y == 2, 0], X[pred_y == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
```

Plotting the centroids of the clusters

```python
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()

plt.show()
```
